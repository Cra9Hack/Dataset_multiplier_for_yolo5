```python
import os
from skimage import color, io, measure
from PIL import Image
from itertools import product

def tile_label(filename, dir_in, dir_out, d,label_list):
    name, ext = os.path.splitext(filename)
    img = Image.open(os.path.join(dir_in, filename))
    w, h = img.size
    grid = product(range(0, h-h%d+1, d), range(0, w-w%d+1, d))
    
    for i, j in grid:
        box = (j, i, j+d, i+d)
        out = os.path.join(dir_out, f'{name}_{i}_{j}{ext}')
        label_out =f'{name}_{i}_{j}'+'.txt'
        f = open(label_out,"w")
        print("file_open")
        for line in label_list:
            if i+d> float(line[2])*h > i and j+d> float(line[1])*w>j:
                f.write(' '.join(normalize_label(line,d,w,h))+'\n') 
        f.close()
        print("file_close")
        img.crop(box).save(out)
        
        
def normalize_label(label,slice_size,full_size_x,full_size_y):
    mltp_x = int(full_size_x/slice_size)
    mltp_y = int(full_size_y/slice_size)
    obj_class = label[0]
    x=float(label[1])*full_size_x
    if x>int(x):
        x=int(x)+1
    l = [ (slice_size*x)/full_size_x for x in range(1,mltp_x)]
    l.reverse()
    for k in [full_size_x*x for x in l]:
        if x>k:
            x = x-k
    x = x/slice_size*1.
    y = float(label[2])*full_size_y
    if y>int(y):
        y=int(y)+1
    l = [ (slice_size*y)/full_size_y for y in range(1,mltp_y)]
    l.reverse()
    for k in [full_size_y*y for y in l]:
        if y>k:
            y = y-k
    y = y/slice_size*1.
    w=mltp_x*float(label[3])
    h=mltp_y*float(label[4])
    return [obj_class, str(round(x,6)),str(round(y,6)),str(w),str(h)]

labels = []
image_set = io.imread_collection('*.jpg')
"""
for file in os.listdir(os.getcwd()):
    if file.endswith(".txt"):
        labels.append(file)
        print(file)
"""
for i, image in enumerate(image_set):
    file = open(image_set.files[i][:-4]+".txt","r")
    label_list = []
    for line in file:
        label_list.append(str.split(line.rstrip(),' '))
    file.close()
    tile_label(image_set.files[i][:-4]+'.jpg',"","",1856,label_list)

```
